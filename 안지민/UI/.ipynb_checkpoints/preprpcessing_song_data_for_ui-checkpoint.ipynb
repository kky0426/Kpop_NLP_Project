### 2-3. 반대되는 분위기 pair 제거
1. 설레는 <-> 나쁜 : 0개
2. 분위기 좋은 <-> 어두운, 나쁜 (어두운 243개, 나쁜 0개)
- [분위기좋은, 어두운, 나쁜]이 같이 있을 경우 분위기 선택 기준
- 설레는, 신나는이 같이 있으면 '어두운,나쁜'을 드랍
- 나머지의 경우 '분위기좋은'을 드랍
3. 잔잔한 <-> 파워풀한, 신나는, 댄스 : 1399개
- [잔잔한, 파워풀한, 신나는, 댄스]가 같이 있을 경우 분위기 선택 기준
- 장르 기준으로 제거하기: 발라드일 경우 '잔잔한' 외 다 드랍/ 댄스일 경우 '잔잔한' 드랍
- 이 경우 다른 장르, 랩/힙랍/록/메탈 등은 드랍기준을 세우기 어려움
4. 어두운 <-> 설레는, 신나는
- [어두운, 설레는, 신나는]이 같이 있을 경우 분위기 선택 기준
- 이별이 같이 있을 경우 '설레는' 드랍
- 신나는이 같이 있을 경우 '어두운' 드랍
